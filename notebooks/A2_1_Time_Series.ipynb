# Part 1: One time-series / bar chart

### Data Loading

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Loading the data 
data_path = os.path.abspath(os.path.join(os.pardir, "data"))
cleaned_data_path = os.path.join(data_path, "Police_Department_Incident_Reports_Complete.csv")
df = pd.read_csv(cleaned_data_path)

# Removing 2025 data
df = df[df['Year'] != 2025]

df_drug = df[df['Category'] == 'DRUG/NARCOTIC'].copy()
```

### Time series 

```python
import matplotlib.pyplot as plt

# Year-by-year aggregation: count crimes per year
yearly_counts = df_drug.groupby('Year').size().reset_index(name='Count')

# Create the time series bar chart
plt.figure(figsize=(10, 6))
plt.bar(yearly_counts['Year'], yearly_counts['Count'], color='#9ecae1', edgecolor='#2171b5')
plt.title('Drug/Narcotic Crimes by Year in San Francisco')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(yearly_counts['Year'], rotation=45)
plt.tight_layout()

# Save the figure to a file
plt.savefig("../plots/drug_crimes_by_year.png", dpi=300)

# Display the figure in the notebook
plt.show()
```
